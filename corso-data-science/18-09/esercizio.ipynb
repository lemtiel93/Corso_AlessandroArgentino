```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt

lista_citta = ['Milano','Napoli','Agrate Brianza','Roma', 'Torino']

n = 30

citta = random.choices(lista_citta, k=n)

temperature = np.random.randint(0,40, size = n).astype(float)




df = pd.DataFrame({
    'Città': citta,
    'Temperatura': temperature,
})
```

```python
temp_max = df['Temperatura'].max()
temp_min = df['Temperatura'].min()
temp_mean = df['Temperatura'].mean()
temp_median = df['Temperatura'].median()
```

```python
categories = [f'Max {df['Città'][df['Temperatura'].idxmax()]}', f'Min {df['Città'][df['Temperatura'].idxmin()]}', 'Media', 'Mediana']
values = [temp_max,temp_min,temp_mean,temp_median]

plt.figure()
plt.bar(categories, values)
plt.title('Grafico a barre')
plt.xlabel('Categorie')
plt.ylabel('Valori')
plt.show()
```

```python
plt.figure()
plt.plot(df.index,df['Temperatura'])
plt.title("Grafico a linea")
plt.xlabel("Temperatura")
plt.ylabel("Giorni")
plt.axhline(temp_mean,color = "red",label = "Media")
plt.legend()
plt.show()
```

```python
plt.figure()
plt.scatter(df.index,df['Temperatura'])
max_index = df['Temperatura'].idxmax()

plt.scatter(max_index, temp_max, color='red', s=100, label='Temperatura Massima', edgecolor='black')
min_index = df['Temperatura'].idxmin()
plt.scatter(min_index, temp_min, color='green', s=100, label='Temperatura Minima', edgecolor='black')
plt.title("Scatter plot")
plt.xlabel("Giorni")
plt.ylabel("Temperatura")
plt.legend()
plt.show()
```
