```python
import pandas as pd
import numpy as np
import random

lista_citta = ['Milano','Napoli','Agrate Brianza']
lista_prodotto = ['Xbox','PS5','Switch']

n = 30

citta = random.choices(lista_citta, k=n)
prodotti = random.choices(lista_prodotto, k=n)
vendite = np.random.randint(1,5, size = n).astype(float)

index = np.random.randint(0,n, size = 3)
for i in index:
    vendite[i] = np.nan

df = pd.DataFrame({
    'Città': citta,
    'Prodotti': prodotti,
    'Unità vendute': vendite
})
df_nuovo = df.dropna()
df_nuovo
```

```python
df.pivot_table(values = 'Unità vendute', columns= 'Città', index = 'Prodotti', aggfunc = 'mean')
```

```python
print(df.groupby('Prodotti')[['Unità vendute']].sum())
```
