```python
import pandas as pd

clienti = []
for i in range(10):
    num_cliente = i
    nome = input("Nome: ")
    cognome = input("Cognome: ")
    data_nascita = input("Data di nascita: ")
    regione = input("Regione: ")
    
    clienti.append([num_cliente, nome, cognome, data_nascita, regione])
```

```python
df_clienti = pd.DataFrame(clienti, columns=["numero", "nome", "cognome", "data", "regione"])
df_clienti
```

```python
df_clienti.to_csv("clienti.csv", index= False)
```

```python
import datetime

today = datetime.datetime.now()
print(today)
df_clienti
```

```python
df_clienti['data'] = pd.to_datetime(df_clienti['data'],format ='%d-%m-%Y')
```

```python
today - df_clienti['data'][0]
```

```python
def calcola_eta(data):
    eta = today.year - data.year
    if today.month < data.month:
        eta = eta - 1
    elif today.month == data.month and today.day < data.day:
        eta = eta - 1
    return eta
```

```python
df_clienti['eta'] = df_clienti['data'].apply(calcola_eta)
```

```python
df_clienti['17'] = df_clienti['eta'].apply(lambda x: 'X' if x > 17 else '')
df_clienti['20'] = df_clienti['eta'].apply(lambda x: 'X' if x > 20 else '')
df_clienti['minorenni'] = df_clienti['eta'].apply(lambda x: 'X' if x < 18 else '')
df_clienti
```

```python
df_clienti.to_csv("clienti2.csv", index = False)
```
